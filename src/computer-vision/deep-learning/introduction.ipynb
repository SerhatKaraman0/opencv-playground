# Perceptron
- Perceptron is the simplest piece of neural network can be used for simple binary classification tasks.
- It has an Input Layer, can have hidden layer (if it is not single layer if it is not single layer it is simply ANN) and an output layer 
- takes the inputs from input layer multiply inputs with weights and an additional bias is added.
- After that it is fed into an activation function which makes the final decision in our case which class the data belongs to (binary classification task).
- linear seperable problems can be solved using the single layer perceptron.



# MultiLayer Perceptron Model

## Forward Propagation
- Applying z = Sum(x0*w0, x1*w1, x2*w2, .., xn*wn) + b to every neuron in the hidden layer.
- Activation_Func(z) => Activation function can be step_func, sigmoid, relu, softmax etc.
- When we get to the output layer we calculate the *loss_function* 
- Output of the loss function gives the error => (y - y_pred)
- Then we calculate the cost by Cost Function => Sum((y-y_pred)^2)

## Backpropagation and Weight Update 
- w_new = w_old - learning_rate
- Role of optimizer: reduce the loss value => by updating weights 
- learning rate should be a small value for not to miss the global minima.




## Vanishing Gradient Problem 
- When w_new is approximately equal to w_old 
- So problem is weights are not updating which causes model to not learn and stay at the same place
- This problem is shown in the very deep neural networks 

# Activation Functions 
## Sigmoid Function
- Suitable for Binary Classification Tasks
- Gives an output between 0 and 1 
- PROBLEM: Derrivative of the Sigmoid Function output is always between the values 0 and 0.25 => this can cause Vanishing Gradient Problem in very deep neural networks in the backpropagation phase
- PROBLEM: Function output is not zero centered. (Being zero centered makes the function efficient in weight updating)
- PROBLEM: Time consuming 

## tanh Function
- Outputs between -1 and 1
- Derrivative between 0 and 1
- Function is 0 centric
- PROBLEM: In a very deep neural network may cause Vanishing Gradient Problem
- PROBLEM: SLow in computation, Time consuming.

## ReLU Function => max(0, x)
- Solves Gradient Vanishing problem and very popular.
- PROBLEM: Creates Dead Neurons if output of the ReLU function is 0.
- PROBLEM: Not 0 centric 

## Leaky ReLU => max(0.01x, x) and Parametric ReLU Function => max(alpha * x, x)
- Same like ReLU except solves dead neuron problem.
- PROBLEM: Not 0 centric

## ELU Activation Function
- No Dead Neuron issue
- Zero centric
- PROBLEM: Slower in computation. Time Consuming 

## Softmax Activation Funcion 
- Used for Multiclass classification problem in the output layer.

# Where to use which activation function
- Hidden Layers => ReLU or its variants 
- Output Layers => Sigmoid (Binary Classification) or Softmax (Multiclass Classification) depending on the classification task we have 


# Loss Functions and Cost Functions 
- Loss Function is calculated for every data point we have 
- For the cost function all data points are passed once then we reduce the cost function by some optimizers 

# Loss Functions for Regression Tasks
- Mean Squared Error  => 1/n * sum((y-y_pred)^2) => Good for regression problems, if you have outlier data dont use it since the outlier incereases the bias too much.
- Mean Absolute Error => 1/n * sum(abs(y-y_pred)) => Better for data with outlier values.
# Loss Functions for Classification Tasks
- Binary Cross Entropy (Binary Class)           
- Categorical Cross Entropy (Multiple Class)  
- Sparse Cross Entropy (Multiple Class)

## Optmizers
- Optimizers are responsible for reducing the loss function in the backpropagation.

## Gradient Descent Optimizers 
- There are two important topics when it comes to Gradient Descent Optimizers 
- Epochs and Iteration
- Convergence will happen 
- PROBLEM: Computation Intensive, Ram and Cpu usage is high 

## Stochastic Gradient Descent Optimizer 
- 1 Iteration for every data point inside 1 epoch
- Solves resource problem of GDO
- PROBLEM: time complexity 
- PROBLEM: convergence will take time 
- PROBLEM: noise gets introduced 

## Mini Batch with SGD
- Convergence speed is increased
- Less noise than SGD
- Efficient resource usage 
- PROBLEM: noise still exist 

## SGD with Momentum
- *Exponential Weight Average* 
- Reduces the noise
- Convergence to global minima

## AdaGrad
- Dynamic Learning Rate 
- But may have Gradient Vanishing Problem 

## RMSPROP
- Exponential Weighted Average 
- Dynamic Learning Rate + Smoothening 

## Adam Optimizer 
- Features of SGD with Momentum + Dynamic Learning Rate + Smoothening   

## Exploding Gradient Problem
- If the initialized weight is very big then during the iterations we move away from the global minima.

# Dropout Layer 
- If our model perform well on training data but not in test data then our model is overfitting.
- In the dropout layer p percentage of neurons are deactivated. 0 <= p <= 1
